```python
## Objective
"""The objective of this project is to analyze sales transactions
to understand sales performance, pricing strategy, discount impact,
and regional trends using data analysis and visualization."""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("../data/sales_data.csv")
df.head()
```

```python
df.info
df.describe
df.isnull().sum()
```

```python
df['Total_Cost'] = df['Unit_Cost'] * df['Quantity_Sold']
df['Revenue'] = df['Unit_Price'] * df['Quantity_Sold']
df['Profit'] = df['Revenue'] - df['Total_Cost']
df.head()
```

```python
## Research Questions
"""1. Which product categories generate the highest profit?
2. How do discounts affect profit and sales amount?
3. Which regions perform best in terms of revenue?
4. Does higher quantity sold always mean higher profit?"""
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x="Product_Category",
    y="Profit",
    data=df,
    estimator=sum
)
plt.title("Total Profit by Product Category")
plt.xticks(rotation=45)
plt.show()
plt.savefig("../outputs/figures/profit_by_category.png")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Discount",
    y="Profit",
    data=df
)
plt.title("Discount vs Profit")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x="Region",
    y="Revenue",
    data=df,
    estimator=sum
)
plt.title("Total Revenue by Region")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Quantity_Sold",
    y="Profit",
    data=df
)
plt.title("Quantity Sold vs Profit")
plt.show()
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

df.groupby('Sale_Date')['Sales_Amount'].sum().plot(figsize=(10,5))
plt.title("Sales Trend Over Time")
plt.ylabel("Sales Amount")
plt.show()
```
